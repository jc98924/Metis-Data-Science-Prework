```python
# Question 1
import csv
import re
from collections import defaultdict
faculty_path = 'faculty.csv'


def count_degrees(csv_file_name):

    """
    Function to open a the faculty .csv file, clean the degree' column, and 
    output a dictionary object where the key is the degree and the value is
    the number of occurences of each degree
    ---
    args: faculty_path = 'faculty.csv'
    output: degree_counter, defaultdict object {key=degree, val = occurences}
    """
 
    with open(faculty_path, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        csv_reader.__next__() 
        output_list = [ ]
        degree_counter = defaultdict(int)
        
        degree_list = [row[1].strip() for row in csv_reader]     
        
        for entry in degree_list:
            if len(entry)>5:
                output_list.extend(re.split('\s+', entry))
            else:
                output_list.append(entry)
        
        cleaned_list = [(re.sub('\.+','',degree)) for degree in output_list]

        for deg in cleaned_list:
            degree_counter[deg] += 1
        return(degree_counter)   

count_degrees(faculty_path)
```

```python
# Question 2, Different Outputs
import csv
import re
from collections import defaultdict

faculty_path = 'faculty.csv'

def count_titles(csv_file_name):
    """
    Function to open a the faculty .csv file, clean the title' column, and 
    output a dictionary object where the key is the title and the value is
    the number of occurences of each title
    ---
    args: faculty_path = 'faculty.csv'
    output: title_counter, defaultdict object {key=title, val = occurences}    
    """
    with open(faculty_path, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        header = csv_reader.__next__() 

        title_list = []
        cleaned_title_list = []
        title_count = defaultdict(int)
    
        title_list = [row[2] for row in csv_reader]
    
        for entry in (title_list):
            s = re.search('\w+',entry)
            cleaned_title_list.append(s.group().lower())
            

        for title in cleaned_title_list:
            title_count[title] += 1
        return title_count
count_titles(faculty_path)
```

```python
# Question 4:

import re
def unique_domains(emails):
    """
    Function take a list of emails and output any instance of a unique domain name
    ---
    args: emails = 'input.txt'
    output: unique_domains,  list object [unique_domains]    
    """
    pattern = r'@([a-z_.]+)'
    domain_list = []
    for email in emails:
        domain_list.append(re.findall(pattern, email)[0])

    return list(set(domain_list))

emails = open('input.txt', 'r')
emails = [email.strip() for email in emails]
 
domains = unique_domains(emails)
for domain in sorted(domains):
    print(domain)
```

Write a function which reads in "faculty.csv", and then returns a dictionary where keys are tuples of names, and values are corresponding rows. You can assume that the tuples are unique, so there iwll be a unique row per tuple.

```python
"""
with open('faculty.csv', 'w') as f:
    f.write(facultycsv)
"""


import csv
import re

def get_dict():
    
    """    
    Function which takes the 'faculty.csv' files and outputs a dictionary with keys
    as tuples based on full names with values as the corresponding rows.
    ---
    args: 'faculty.csv'
    return: dictionary datatype: {key: tuple(first,middle,last), value: [rows[1:4]]}
    """ 
    with open('faculty.csv', 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        header = csv_reader.__next__()
        
        name_dict = {}
        regex = re.compile('\s')
        
        for row in csv_reader:
            name_dict[tuple(regex.split(row[0]))] = row[1:4]
        return name_dict    

get_dict()

"""
HackerRank Validation 
answer = get_dict()
for key, val in answer.items():
    assert '{key},{val}'.format(key=' '.join(key), val=','.join(val)) in facultycsv
assert len(answer) == facultycsv.count('\n')
print(1)
"""
```
